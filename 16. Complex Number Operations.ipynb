```python
def getting_the_number():
    print("A complex number is in the format of (a +bj).")
    while True:
        
        try:
            real_part = float(input("Enter the real part of the complex number(a): "))
            break
        except:
            print("Please enter a valid number.")
    while True:
        try:
            imag_part = float(input("Enter the imaginary part of the complex number(b): "))
            break
        except:
            print("Please enter a valid number.")
    
    z = (real_part) + (imag_part)*1j
    return z,real_part,imag_part
```

```python
class ComplexOperations:
        def __init__(self,r1,r2,i1,i2):
            
            self.r1 = r1
            self.i1 = i1
            self.z1 = r1 + i1*1j
            self.r2 = r2
            self.i2 = i2
            self.z2 = r2 + i2*1j
 
        def add(self):
            return self.z1 + self.z2
        
        def minus(self):
            return f'''{self.z1 - self.z2} is z1 -z2  
{self.z2 - self.z1} is z2 - z1)'''
        
        def multiply(self):
            return self.z1 * self.z2
        
        def divide(self):
            part1 = (self.r1*self.r2 + self.i1*self.i2)/(self.r1**2 + self.i2**2) 
            part2 = (self.i1*self.r2 - self.r1*self.i2)/(self.r1**2 + self.i2**2)*1j
            return part1 + part2
        
        def conjugate(self):
            conj1 = f'Conjugate of {self.z1} is {self.z1.conjugate()}'
            conj2 = f'Conjugate of {self.z2} is {self.z2.conjugate()}'
            return conj1 + "\n" + conj2
```
