```python
def getting_number():
    while True:
        try:
            num = int(input("Please enter a number: "))
            break
        except:
            print("Error")
    return num
```

```python
def naming_number(n):
    if len(str(n)) == 1:
        if n == 1:
            return "One"
        elif n == 2:
            return "Two"
        elif n == 3:
            return "Three"
        elif n == 4:
            return "Four"
        elif n == 5:
            return "Five"
        elif n == 6:
            return "Six"
        elif n == 7:
            return "Seven"
        elif n == 8:
            return "Eight"
        elif n == 9:
            return "Nine"
        else:
            return "Zero"
#########################################################################################################################
# Two digits divisible by 10
    if len(str(n)) == 2:
        if n == 10:
            return "Ten"
        if n == 20:
            return "Twenty"
        if n == 30:
            return "Thirty"
        if n == 40:
            return "Forty"
        if n == 50:
            return "Fifty"
        if n == 60:
            return "Sixty"
        if n == 70:
            return "Seventy"
        if n == 80:
            return "Eighty"
        if n == 90:
            return "Ninety"
###########################################################################################################################
#two digits not divisible by ten     
#first digit is 1
        if n // 10 == 1: 
            if n % 10 == 1:
                return 'Eleven'
            elif n % 10 == 2:
                return 'Twelve'
            elif n % 10 == 3:
                return 'Thirteen'
            elif n % 10 == 4:
                return 'Fourteen'
            elif n % 10 == 5:
                return 'Fifteen'
            elif n % 10 == 6:
                return 'Sixteen'
            elif n % 10 == 7:
                return 'Seventeen'
            elif n % 10 == 8:
                return 'Eighteen'
            elif n % 10 == 9:
                return 'Nineteen'
#first digit other than 1
        else:
            return two_digits_concat(n)
######################################################################################################################
    if len(str(n)) == 3:
        return three_digits_concat(n)
       
    if len(str(n)) == 4:
        return four_digit_concat(n)
    
    if len(str(n)) == 5:
        return five_digit_concat(n)
    
    if len(str(n)) ==6:
        return six_digit_concat(n)
    
    if len(str(n)) ==7:
        return seven_digit_concat(n)
    
    if len(str(n)) ==8:
        return eight_digit_concat(n)
    
    if len(str(n)) == 9:
        return nine_digit_concat(n)
    if len(str(n)) == 10:
        return billion(n)
    if len(str(n)) > 10:
        print("Nah fuck that.")
```

```python
def two_digits_concat(n):
    
    for i in range(2,9+1):
            if n //10 == i:
                first_part = naming_number(int(str(i)+'0'))
                for i in range(1,9+1):
                    if n % 10 == i:
                        second_part =" "+ naming_number(i)
    return (first_part + second_part)
        
    
def three_digits_concat(n):
    
    for i in range (1,10):
        if n%100 == 0:
            if n // 100 == i:
                return naming_number(i) + ' Hundred'
    
        elif n // 100 == i:
            first_part = naming_number(i) + ' Hundred'
            second_part = naming_number(int(str(n)[1:]))
            return first_part + " " + second_part

def four_digit_concat(n):
    for i in range(1,10):
        if n % 1000 == 0:
            if n // 1000 == i:
                return naming_number(i) + " Thousand"
        elif n // 1000 == i:
            first_part = naming_number(i) + " Thousand"
            second_part = naming_number(int(str(n)[1:]))
            return first_part +" "+ second_part
        
def five_digit_concat(n):
    for i in range(10,99+1):
        if n % 1000 ==0:
            if n // 1000 == i:
                return naming_number(i) +" Thousand"
        elif n // 1000 == i:
            first_part = naming_number(i) + " Thousand"
            second_part = naming_number(int(str(n)[2:]))
            return first_part + " " + second_part
        
def six_digit_concat(n):
    for i in range(100,999+1):
        if n % 1000 == 0:
            if n // 1000 == i:
                return naming_number(i) + " Thousand"
        elif n // 1000 == i:
            first_part = naming_number(i) + " Thousand"
            second_part = naming_number(int(str(n)[3:]))
            return first_part + ' ' + second_part
            
def seven_digit_concat(n) :
    for i in range(1,9+1):
        if n % 10**6 == 0:
            if n // 10**6 == i:
                return naming_number(i) + " Million"
        elif n // 10**6 == i:
            first_part = naming_number(i) + " Million"
            second_part = naming_number(int(str(n)[1:]))
            return first_part + ' ' + second_part
        
def eight_digit_concat(n) :
    for i in range(10,99+1):
        if n % 10**6 == 0:
            if n // 10**6 == i:
                return naming_number(i) + " Million"
        elif n // 10**6 == i:
            first_part = naming_number(i) + " Million"
            second_part = naming_number(int(str(n)[2:]))
            return first_part + ' ' + second_part

def nine_digit_concat(n) :
    for i in range(100,999+1):
        if n % 10**6 == 0:
            if n // 10**6 == i:
                return naming_number(i) + " Million"
        elif n // 10**6 == i:
            first_part = naming_number(i) + " Million"
            second_part = naming_number(int(str(n)[3:]))
            return first_part + ' ' + second_part
        
def billion(n):
    for i in range(1,9+1):
        if n % 10**9 == 0:
            if n // 10**9 == i:
                return naming_number(i) + " Billion"
        elif n // 10**9 == i:
            first_part = naming_number(i) + " Billion"
            second_part = naming_number(int(str(n)[1:]))
            return first_part + ' ' + second_part
    
```
