```python
import matplotlib.pyplot as plot
```

```python
ktm_y = [1_500,2_100,3_000]
ktm_x = ["April","May","June"]

brt_y = [1000,2000,2300]
brt_x = ["April","May","June"]

```

```python
plot.plot(ktm_x,ktm_y,label  = "KTM") #plotting the line on the graph
plot.plot(brt_x,brt_y,label = "BRT") # another line
plot.xlabel("Months") # the naming of x axis
plot.ylabel("Patients") # naming of y axis
plot.title("Covid data") # naming of the whole graph
plot.legend() # displays the legend if you have given label to the lines
```
