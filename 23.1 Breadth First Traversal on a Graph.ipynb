```python
graph = {
    'a':['b','c'],
    'b':['d'],
    'c':['e'],
    'd':['f'],
    'e':[],
    'f':[]
}

```

```python
def BFS(graph,startnode):
    queue = collections.deque([startnode]) # creating a queue with the collections module
                                           # it adds the start node to the queue object and declares its type as a queue 
    while queue: # while it still has elements
        current = queue.popleft()
        print(current)
        for neighbor in graph[current]:
            queue.append(neighbor)
```

```python
BFS(graph,'a')
```

```python
BFS(graph,'c')
```
