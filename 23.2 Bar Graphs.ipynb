```python
import matplotlib.pyplot as bar
```

```python
#xaxis = ["Jan",'Feb','Mar','Apr','May']
xaxis = [1,2,3,4,5]
yaxis = [250,600,450,200,350]
bar_label = ["Jan",'Feb','Mar','Apr','May']
bar.xlabel('Months')
bar.ylabel("Active Cases")
bar.title("Covid Data")
bar.bar(xaxis,yaxis,
        tick_label = bar_label,
        align = 'center',
        color = ['m','g'])
```
