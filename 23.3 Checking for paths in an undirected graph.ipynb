```python
edges = [
    ('i','j'),
    ('k','i'),
    ('m','k'),
    ('k','l'),
    ('o','n')
]
```

```python
#creating an undirected graph adjacency list with the help of edges

graph = {}
for first,second in edges:
    if first in graph:
        graph[first].append(second)
    else:
        graph[first] = [second]
        
    if second in graph:
        graph[second].append(first)
    else:
        graph[second] = [first]
```

```python
graph
```

```python
def has_path_DFS(graph,start,end,visited = set()):
    if start == end:
        return True
    
    visited.add(start)
    
    for neighbor in graph[start]:
            if neighbor not in visited:
                visited.add(neighbor)
                
                if has_path_DFS(graph,neighbor,end,visited):
                        return True
    
    return False
```
