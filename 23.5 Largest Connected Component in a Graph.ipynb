```python
graph = {
    1:[0],
    2:[3,4],
    3:[2,4],
    4:[2,3],
    5:[0,8],
    0:[1,5,8],
    8:[0,5]
}
```

```python
# the one above has 4 nodes which is obv the largest
# the one below has 3 
# we need to find a program that returns 4
```

```python
def exploring(graph,start,visited):
    if start in visited:
        return 0
    
    size = 1 #1 represents the current node that im at and it accumulates over time
    
    visited.add(start)
    
    for neighbor in graph[start]:
        size += exploring(graph,neighbor,visited)
        
        
    return size
```

```python
def normalDFS(graph,start,visited,nodes):
    if start in visited:
        return
    
    visited.add(start)
    nodes.append(start)
    
    
 
    for neighbor in graph[start]:
       
        normalDFS(graph,neighbor,visited,nodes)
        
        
    return nodes
```

```python
def largest_result(graph):
    visited = set()
    largest = 0
    largest_node = ''
    for node in graph:
        size = exploring(graph,node,visited)
        if size >= largest:
            largest = size
            largest_node = node
    return largest,largest_node

        
        
```

```python
def result_printing(graph):
    largest,largest_node = largest_result(graph)
    visited = set()
    nodes = []
    print(f'The largest connected component in the graph has {largest} nodes. They are: \n')
    for item in normalDFS(graph,largest_node,visited,nodes):
        print(f'{item} : {graph[item]}')
```

```python
result_printing(graph)
```
