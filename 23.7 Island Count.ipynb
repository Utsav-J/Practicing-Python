```python
# here the blue squares are water and yellow are land
# we need to find how many pieces of land there are
# above there are 4 pieces of land
```

```python
# we are gonna have to go thru every square
# if that square is a water square, we gon do nothing
# but if that square is a land, we gon do a DFS on that square and mark all the other squares we visit during that process
# we keep count for every successful DFS we do, THAT WILL BE OUR ISLAND COUNT
# we go on further and explore new pieces of land
# if we come thru a previously seen piece of land, we're not gonna do a DFS on it and just keep going
# if we get a land square and it also hasnt been visited yet, we gonna do a DFS on it and increment the island count
```

```python
grid = [
    ['W','L','W','W','L','W'],
    ['L','L','W','W','L','W'],
    ['W','L','W','W','W','W'],
    ['W','W','W','L','L','W'],
    ['W','L','W','L','L','W'],
    ['W','W','W','W','W','W'],
]
```

```python
def explore(grid,row,coln,visited):
    position = str(row) + ',' + str(coln)
    
    if position in visited:
        return False                    #avoiding cycles
    visited.add(position)
    
    row_inbounds = row in range(0,len(grid))
    coln_inbounds = coln in range(0,len(grid[0]))        #avoiding out of bounds row and coln number
    
    if not (row_inbounds and coln_inbounds):
        return False
    
    if grid[row][coln] == "W":          #avoiding water squares
        return False
    
    explore(grid,row - 1,coln,visited) #up
    explore(grid,row + 1,coln,visited) #down
    explore(grid,row,coln - 1,visited) #left
    explore(grid,row,coln + 1,visited) #right
    
    return True # THIS TRUE WILL ONLY BE EXECUTED IF ITS A LAND AND IT HASNT BEEN VISTED BEFORE, SO IT DENOTES A SUCCESFULL 
                 # TRAVERSAL FROM A CERTAIN POINT IN THE GRAPH
```

```python
def island_count(grid):
    visited = set()
    count = 0
    for row in range(0,len(grid)):
        for coln in range(0,len(grid[0])):
            if explore(grid,row,coln,visited) == True:
                count += 1
    
    
    return count
```

```python
island_count(grid)
```
