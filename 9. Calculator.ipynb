```python
def sine():
    degrees_or_radian = input("Are you entering the value in degrees or radian?")
    
    if degrees_or_radian[0].upper() == "D":
        while True:
            try:
                num = float(input('Please enter the degrees value: '))
                num = math.radians(num)
                break
            except:
                print("Please only enter the number without letters or punctuation! ")
                
    elif degrees_or_radian[0].upper() == "R":
        while True:
            try:
                num = float(input('Please enter the radians value: '))
                break
            except:
                print("Please only enter the number without letters or punctuation! ")
    else:
        print("Dont mess w me")
    
    return num, math.sin(num)

def cosine():
    degrees_or_radian = input("Are you entering the value in degrees or radian?")
    
    if degrees_or_radian[0].upper() == "D":
        while True:
            try:
                num = float(input('Please enter the degrees value: '))
                num = math.radians(num)
                break
            except:
                print("Please only enter the number without letters or punctuation! ")
                
    elif degrees_or_radian[0].upper() == "R":
        while True:
            try:
                num = float(input('Please enter the radians value: '))
                break
            except:
                print("Please only enter the number without letters or punctuation! ")
    else:
        print("Dont mess w me")
    
    return num, math.cos(num)


def tantan():
    degrees_or_radian = input("Are you entering the value in degrees or radian?")
    
    if degrees_or_radian[0].upper() == "D":
        while True:
            try:
                num = float(input('Please enter the degrees value: '))
                num = math.radians(num)
                break
            except:
                print("Please only enter the number without letters or punctuation! ")
                
    elif degrees_or_radian[0].upper() == "R":
        while True:
            try:
                num = float(input('Please enter the radians value: '))
                break
            except:
                print("Please only enter the number without letters or punctuation! ")
    else:
        print("Dont mess w me")
    
    return num, math.tan(num)
    
```

```python
def square_root():
    while True:
        try:
            num = float(input('Please enter the number you wanna calculate a square root of:  '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
    
    return num,math.sqrt(num)

def exponent():
    while True:
        try:
            num1 = float(input('Please enter the base number: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
            
    while True:
        try:
            num2 = float(input('Please enter power for the base number: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
            
    return num1,num2, math.pow(num1,num2)
```

```python
def adding():
    while True:
        try:
            num1 = float(input('Please enter the first number you wanna add: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
            
    while True:
        try:
            num2 = float(input('Please enter the second number you wanna add: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
    
    
    return num1,num2, (num1 + num2)
            
```

```python
def minus():
    while True:
        try:
            num1 = float(input('Please enter the first number you wanna subtract from: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
            
    while True:
        try:
            num2 = float(input('Please enter the second number you wanna subtract: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
            
            
                

    return num1,num2, (num1 - num2)
            
```

```python
def multiplying():
    while True:
        try:
            num1 = float(input('Please enter the first number you wanna multiply: '))
            
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
    
    while True:
        try:
            num2 = float(input('Please enter the second number you wanna multiply: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
    
    return num1,num2, (num1 * num2)
            
```

```python
def dividing():
    while True:
        try:
            num1 = float(input('Please enter the first number you wanna divide from: '))
            
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
            
    while True:
        try:
            num2 = float(input('Please enter the second number you wanna divide with: '))
            break
        except:
            print("Please only enter the numbers without letters or punctuation! ")
    
    return num1,num2, (num1 / num2)
            
```
